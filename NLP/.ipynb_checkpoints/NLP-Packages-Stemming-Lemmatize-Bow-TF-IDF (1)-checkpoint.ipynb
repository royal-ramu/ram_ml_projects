# Stemming

```python
#1.Porter Stemming
#2.Snowball Stemming
#3.Lancaster Stemming
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
words = ['sincerely','electricity','roughly','ringing']
porter = PorterStemmer()
for w in words:
    print(w, " : ", porter.stem(w))
```

```python
from nltk.stem.snowball import SnowballStemmer
snow = SnowballStemmer(language='english')
for w in words:
    print(w, " : ", snow.stem(w))
```

```python
from nltk.stem.lancaster import LancasterStemmer
lan=LancasterStemmer()
for w in words:
    print(w, " : ", lan.stem(w))
```

```python
from nltk.stem.lancaster import LancasterStemmer
lan=LancasterStemmer()
text='It has become appallingly obvious  that our technology has exceeded our humanity.'
```

```python
#using function - list comprehension

from nltk.stem.lancaster import LancasterStemmer
lan=LancasterStemmer()
def tokenizer_lancaster(text):
            return [lan.stem(word) for word in text.split()]
tokenizer_lancaster('It has become appallingly obvious  that our technology has exceeded our humanity.')
```

# LEMMATIZATION

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
paragraph = """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               """
```

```python
sentences = nltk.sent_tokenize(paragraph)
print(sentences)
```

```python
import nltk
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
# Lemmatization
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```

```python
print(sentences)
```

# Bow

```python
# Bow-Bag of words
list1=["Dog bites man.","Man bites dog.","Dog eats meat.","Man eats food."]
list1
```

```python
#By using list comprehension sentences coverted into
#lower case and . replaced with white space
processed_list1 = [doc.lower().replace(".","") for doc in list1]
processed_list1
```

```python
#look at the  list
print("Our corpus: ", processed_list1)
```

# count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
#Build a BOW representation for the corpus
bow_rep = count_vect.fit_transform(processed_list1)
```

```python
#Look at the vocabulary mapping
print("Our vocabulary: ", count_vect.vocabulary_)
```

```python
#see the BOW rep for the sentences
print("BoW representation for 'dog bites man': ", bow_rep[0].toarray())#1st sentence
print("BoW representation for 'man bites dog: ",bow_rep[1].toarray())#2nd sentence  
print("BoW representation for 'dog eats meat':",bow_rep[2].toarray())
print("BoW representation for 'man eats food':",bow_rep[3].toarray())
# bites dog eats food man meat
   # 0     1    2   3   4   5
```

```python
#Get the representation using this vocabulary, for a new text
temp = count_vect.transform(["dog and dog are friends"])
print("Bow representation for 'dog and dog are friends':", temp.toarray())
```

# Binary Vectorization

```python
#BoW with binary vectors,present =1,not present =0

count_vect = CountVectorizer(binary=True)
count_vect.fit(processed_list1)
temp = count_vect.transform(["dog and dog are friends"])
print("Bow representation for 'dog and dog are friends':", temp.toarray())
```

# TF-IDF

```python
#TF-Term Frequency
#IDF-Inverse Document Frequency
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ['This is the first document.',
   'This document is the second document.',
 'And this is the third one.',
 'Is this the first document?' ]

corpus
```

```python
processed_corpus=[doc.lower().replace(".","") for doc in corpus]
processed_corpus
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_corpus)
```

```python
vectorizer.get_feature_names_out()
```

```python
print('this is the first document',X[0].toarray())
print("this document is the second document",X[1].toarray())
print("and this is the third one",X[2].toarray())
print("is this the first document",X[3].toarray())
```
