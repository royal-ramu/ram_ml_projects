# Preprocessing Text Data

**Wordopt**
---
It is used to clean and preprocess text data before using it in tasks.

It takes a text string as input and applies multiple cleaning steps, such as:

Converting to lowercase

Removing brackets []

Removing special characters and punctuation

Stripping URLs

Removing HTML tags

Eliminating newline characters

Removing words containing numbers

This helps to convert messy, unstructured raw text into a clean, standard format that’s easier to analyze or use in machine learning models.

```python
def wordopt(text): 
    #creating function to process text
    text=text.lower() #Converts all characters in the text to lowercase.
    text=re.sub(r'\[.*?\]','',text) #Removes text inside square brackets.
    text=re.sub("\\W"," ",text) #Replaces all non-word characters (\W) with a space.
    text=re.sub(r'https?://\S+|www\.\S+','',text) #Removes HTML tags, like <div>, <br>, etc.
    text=re.sub('<.*?>+','',text) #multiple tags are removed
    text=re.sub('[%s]'% re.escape(string.punctuation),'',text) #Removes all punctuation marks using the `string.punctuation` set (`!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`).
                                                               #re.escape() ensures that special characters are treated as literals.
    text=re.sub('\n','',text) #Removes newline characters
    text=re.sub(r'\w*\d\w*','',text) # Removes alphanumeric words that contain digits, like `covid19`, `news24`, etc.
    return text #Returns the cleaned and preprocessed version of the input text.
```

```python
df['text']=df['text'].apply(wordopt)
```

# Data Splitting

```python
x=df['text']
y=df['class']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# NLP

```python
vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train) #train-fit
xv_test=vectorization.transform(x_test)#test-transform
```

```python
xv_train
```

```python
xv_train.shape
```

```python
xv_test.shape
```

# Model Building

# Logistic Regression

```python
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr1=LR.predict(xv_test)
pred_lr1
```

```python
LR_acc=LR.score(xv_test,y_test)
LR_acc
```

# Decision Tree Classifier

```python
DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt1=DT.predict(xv_test)
pred_dt1
```

```python
DT_acc=accuracy_score(y_test,pred_dt1)
DT_acc
```

# Random Forest Classifier

```python
RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
```

```python
RF_acc=RF.score(xv_test,y_test)
RF_acc
```

# Performance Analysis on each model

```python
pt=[]
for i in[LR,RF,DT]:
    i.fit(xv_train,y_train)
    pt.append(round(accuracy_score(y_test,i.predict(xv_test))*100,2))
print(pt)
```

# CONCLUSION

From the above algorithms used for modelling the dataset,Decision Tree gives the best accuracy with accuracy score 99.54 %.
